```python
import shap
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from radical_shapley_values import reshape_shapley_output
from radical_shapley_values import compute_shapley_values
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X,y = shap.datasets.adult()
X = X[['Age', 'Hours per week','Education-Num', 'Marital Status', 'Capital Gain', 'Sex']]
nf= 6
```

```python
num_samples = 1000
knn = KNeighborsClassifier()
knn.fit(X, y)
f = lambda x: knn.predict_proba(x)[:,1] # Get the predicted probability that y=True
explainer = shap.KernelExplainer(f, X.iloc[0:100]) # The second argument is the "background" dataset; a size of 100 rows is gently encouraged by the code
kernel_shap = explainer.shap_values(X.iloc[0:num_samples])
```

```python
def shap_payoff_KNN(X, y):
    knn = KNeighborsClassifier()
    knn.fit(X, y)
    return knn.predict_proba(X)[:,1]
```

```python
ypred = shap_payoff_KNN(X,y)
mean_prediction = np.mean(ypred)
radical_shapley = reshape_shapley_output(compute_shapley_values(shap_payoff_KNN, X, y, zero_payoff=np.ones(X.shape[0])*mean_prediction))
radical_shapley = radical_shapley[0:num_samples]
```

```python
plt.figure()
plt.scatter(ypred[0:1000]-mean_prediction, np.sum(radical_shapley, axis=1), s=2)
plt.scatter(ypred[0:1000]-explainer.expected_value, np.sum(kernel_shap, axis=1), s=2)
```

```python
plt.figure(figsize=[9,6])
num_features = X.shape[1]
for feature_idx in range(num_features):
    plt.scatter(radical_shapley[:,feature_idx], kernel_shap[:,feature_idx], s=3)
plt.legend(X.columns,fontsize=14, markerscale=6)
plt.plot([-0.5,0.9],[-0.5,0.9],'k:')
plt.xlabel('Radical Shapley values',fontsize=14)
plt.ylabel('Kernel explainer SHAP values',fontsize=14)
```

```python
shap.summary_plot(kernel_shap,feature_names=X.columns, features=X[0:1000], title='Kernel explainer SHAP Values')
shap.summary_plot(radical_shapley, feature_names=X.columns, features=X[0:1000], title='Radical Shapley Values')
```

```python
# Plot distributions of shapley values
ext = np.max(np.abs(np.concatenate([radical_shapley, kernel_shap])))

plt.figure(figsize=[10,8])
BW = 'Scott'
YL = 15
ax1 = plt.subplot(2,1,1)
for feature_idx in range(nf):
    sns.kdeplot(radical_shapley[:,feature_idx], label=X.columns[feature_idx], shade=True, bw=BW)
plt.xlim([-ext,ext])
plt.ylim([0,YL])
plt.xlabel('Radical Shapley values',fontsize=14)
ax1.legend(X.columns, loc='upper right')

ax2 = plt.subplot(2,1,2)
for feature_idx in range(nf):
    sns.kdeplot(kernel_shap[:,feature_idx], label=X.columns[feature_idx], shade=True, bw=BW)
plt.xlim([-ext,ext])
plt.ylim([0,YL])
ax2.legend(X.columns, loc='upper right')
plt.xlabel('Kernel Explainer SHAP values',fontsize=14)
```

```python
order_diff = np.abs(np.argsort(kernel_shap, axis=1)-np.argsort(radical_shapley, axis=1))
print(np.mean(order_diff,axis=0))
print(np.mean(order_diff)/2)
```
