```python
# Imports
import numpy as np
import pandas as pd
import xgboost as xgb
import shap
import time
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import random

from radical_shapley_values import compute_shapley_values
from radical_shapley_values import shap_payoff_XGBreg as shap_xgb_reg
from radical_shapley_values import shap_payoff_XGBclas as shap_xgb_cls
from radical_shapley_values import reshape_shapley_output
```

```python
# Random dataset
ns = 10000
nf = 3

X = pd.DataFrame(np.random.randn(ns, nf)) # 3 random features
X[[1]] = X[[0]] + np.random.randn(ns,1)*0.2 # x2 is correlated with x1
y = X[[0]] + np.random.randn(ns,1) # y depends on x1 (and so on x2 as well) but not x3
X.columns = ['Strong predictor of y', 'Correlated with strong predictor', 'Not a predictor of y']
```

```python
# Census data dataset
X,y = shap.datasets.adult()
# X = X[['Age', 'Education-Num', 'Marital Status', 'Occupation', 'Sex']]
X = X[['Age', 'Hours per week','Education-Num', 'Marital Status', 'Capital Gain', 'Sex']]
ns = 32561
nf = 6
```

```python
# Change data size
ns = 10000
nf = 6

X = X[X.columns[0:nf]][0:ns]
y = y[0:ns]
```

```python
# XGBoost Classifier
scores = shap_xgb_cls(X,y)
start = time.time()
shapley_values = reshape_shapley_output(compute_shapley_values(shap_xgb_cls, X, y, zero_payoff = np.ones(ns)*np.mean(scores)))
end = time.time()
print('Radical Shapley values run time: %.3f. Per sample: %.3f' %(end-start, (end-start)/ns))


xgb_model = xgb.XGBClassifier(random_state = 1, objective='binary:logistic')
xgb_model.fit(X, y)
explainer = shap.TreeExplainer(xgb_model)
# explainer = shap.TreeExplainer(xgb_model, X[0:100], feature_dependence="independent", model_output="probability")
start = time.time()
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    SHAP_values = explainer.shap_values(X)
end = time.time()
print('Tree SHAP values run time: %.3f. Per sample: %.3f' %(end-start, (end-start)/ns))
```

```python
# XGBoost Regressor
scores = shap_xgb_reg(X,y)
start = time.time()
shapley_values = reshape_shapley_output(compute_shapley_values(shap_xgb_reg, X, y, zero_payoff = np.ones(ns)*np.mean(scores)))
end = time.time()
print('Radical Shapley values run time: %.3f. Per sample: %.3f' %(end-start, (end-start)/ns))

# Get SHAP values
xgb_model = xgb.XGBRegressor(random_state = 1)
xgb_model.fit(X, y)
explainer = shap.TreeExplainer(xgb_model)
start = time.time()
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    SHAP_values = explainer.shap_values(X)
end = time.time()
print('Tree SHAP values run time: %.3f. Per sample: %.3f' %(end-start, (end-start)/ns))
```

```python
# Plot marginal payoffs for feature - run XGBoost Classifier
scores = shap_xgb_cls(X,y)
start = time.time()
shapley_values, payoffs, participation = compute_shapley_values(shap_xgb_cls, X, y, zero_payoff = np.ones(ns)*np.mean(scores))
shapley_values = reshape_shapley_output(shapley_values)
payoffs = reshape_shapley_output(payoffs)
end = time.time()
print('Radical Shapley values run time: %.3f. Per sample: %.3f' %(end-start, (end-start)/ns))
```

```python
# Plot marginal payoffs for feature

feature_idx = 5
num_features = np.log2(payoffs.shape[1]).astype('int')
num_coalitions = payoffs.shape[1]
not_participating = participation[feature_idx]==False
indexes_not_participating = np.where(not_participating)[0]
indexes_participating = indexes_not_participating + 2**(num_features - feature_idx - 1)

payoff_margin = np.zeros([payoffs.shape[0], len(indexes_not_participating)])
for i in range(len(indexes_not_participating)):
    payoff_margin[:,i] = payoffs[:,indexes_participating[i]] - payoffs[:,indexes_not_participating[i]]
other_features = participation[:,indexes_not_participating] # which other features are present in each coalition
other_features = other_features[np.arange(participation.shape[0])!=feature_idx,:] # remove the selected feature from this list

df = pd.DataFrame()
for i in range(int(num_coalitions/2)):
    tmp_df = pd.DataFrame(payoff_margin[:,i], columns = ['margin'])
    tmp_df['num_features'] = np.sum(other_features[:,i])
    tmp_df['coalition_idx'] = i
    tmp_df['feature_val'] = X[X.columns[feature_idx]].values
    for j, feature_name in enumerate(X.columns[np.arange(participation.shape[0])!=feature_idx]):
        tmp_df[feature_name] = other_features[j, i]
    df = pd.concat([df, tmp_df], ignore_index=True)
    
# Downsample data (when num_features is not 0 or all)
df2 = pd.DataFrame()
for i in np.unique(df.num_features):
    tmp_df = df[df.num_features==i]
    tmp_df = tmp_df.sample(n=10000, axis=0)
    df2 = pd.concat([df2, tmp_df], ignore_index=True)
df = df2
```
